```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel('/content/Employee_Data.xls')
df.head()
```

```python
df.drop(['Employee Id', 'First Name', 'Last Name'],axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.displot(df['Age'],color='green')
plt.show()
```

```python
sns.displot(df['Salary'],color='pink')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Experience',y='Salary',color='blue')
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
sns.countplot(x='Department',data=df,hue='Department')
plt.legend(loc='upper right')
plt.show()
```

```python
sns.countplot(x='Age',hue='Age',data=df)
plt.show()
```

```python
sns.barplot(x='Department',y='Salary',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Age',y='Salary',data=df)
plt.show()
```

```python
sns.lmplot(x='Age',y='Salary',data=df)
plt.show()
```

```python
sns.lmplot(x='Age',y='Salary',data=df,hue='Department')
plt.show()
```
